````
# การบวกเริ่มที่ 0 การคูณเริ่มที่ 1
#การหาค่าเฉลี่ยอายุเกิดจากการเอาข้อมูลอายุมาบวกกัน แล้วหารด้วยจำนวนข้อมูลทั้งหมด เริ่มจาก
sum = 0  # ก่อนจะเอาค่ามาบวกกันได้ ต้องสร้างตัวแปรมารับก่อน จึงสร้างตัวแปร sum ขึ้นมา โดยกำหนดให้เท่ากับ 0  และเมื่อมีข้อมูลเข้ามาก็จะทำการบวกเพิ่มเข้าไปในตัวแปร sum เรื่อยๆ
N = 0 # จำนวนข้อมูลทั้งหมด โดยจะนับว่ามีอายุที่เอามาบวกกันกี่ค่า
for a in data_male['age']: # วนลูปค่าอายุแต่ละค่า โดยมาจากคอลัมน์ age
    
    if a > 0: # ทำการตรวจสอบก่อน โดยถ้าอายุมากกว่า 0 คือข้อมูลที่มีค่าปกติ ไม่มีข้อผิดพลาด
        sum += a # หากข้อมูลมีค่ามากกว่า 0 จะนำมาบวกเข้าใน ตัวแปร sum คือ sum + a
        N += 1 #จำนวนข้อมูลที่เข้ามาในแต่ละครั้ง โดยการบวกเข้าที่ละ 1 จนครบ

print(f'อายุเฉลี่ย ของ ผู้ป่วยชาย {sum/N}') #แสดงอายุเฉลี่ยของผู้ป่วยชาย โดยการเอาผลรวมของอายุทั้งหมด หาร จำนวนข้อมูลทั้งหมด
````

## Quiz ตัดตารางออกมาเฉพาะปี 2021 ```announce_date``` ในปี 2021 

**Hint**
*   วนลูปหา index ของปี 2021
*   ตัดตารางมาเฉพาะ ปี 2021



````
TF=list()
for each_row in data_covid.iterrows():
    if each_row[1]['announce_date'].split('/')[2] == '2021':
        TF.append(True)
    else:
        TF.append(False)

data_covid[TF].head()
````
