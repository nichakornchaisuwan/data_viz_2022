```python
data_con4 = os.path.join(path,'confirmed-cases-since-280265.csv') #เอาพาร์ทที่อยู่ในไฟล์มาเชื่อมกับColab
print(data_con4)
data_covid4 = pd.read_csv(data_con4)  # pd.read_csv เป็นการอ่านข้อมูล แล้วตั้งชื่อตัวแปรที่เก็บไว้ว่า data_covid4
data_covid4
```

### แทนค่า Missing ด้วยค่าที่เหมาะสม

#### แทนค่า Missing ของคอลัมน์ Notified data ด้วยค่าที่เหมาะสม

```python
data_covidall=data_covidall.fillna(value ={ 'Notified date' : data_covidall['announce_date']})  ## แก้ค่า Missing ในคอลลัมน์ Notified date โดยใช้คำสั่ง .fillna() ซึ่งค่าที่นำมาใส่คือ ค่าจากคอลลัมน์ announce_date
data_covidall
```

####แทนค่า Missing ของคอลัมน์ *sex* ด้วยจำนวนเท่าๆกัน

```python
data_covidall.info() # เช็ค type ของข้อมูล
```

```python
data_covidall[data_covidall['sex'].isnull()]['No.'].count() # เช็คค่าว่างของคอลัมน์ sex โดยให้นับตามจำนวนคอลัมน์ของลำดับ
```

```python
data_covidall[['sex']] = data_covidall[['sex']].fillna('ชาย',limit=89530) #เติมค่าที่เป็นเพศชาย โดยกำหนด limit ที่ 89530
```

```python
data_covidall[['sex']] = data_covidall[['sex']].fillna('หญิง') # เติมค่าที่เป็น เพศหญิง โดยไม่ได้กำหนดว่า limit ว่าเท่าไหร่ หมายความว่า เติมเพศหญิงลงในค่าว่างที่เหลือทั้งหมด
```

```python
data_covidall[data_covidall['sex'].isnull()]['No.'].count() #เช็คค่าว่างอีกครั้ง
```
